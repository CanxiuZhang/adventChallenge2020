```python
# read input file

input = open('./input.txt', 'r') 
Lines = input.readlines() 
  
count = 0
for line in Lines: 
    print("Line{}: {}".format(count, line.strip())) 
    count += 1
```

```python
# clean data
directions = ['e', 'w', 'se', 'sw', 'ne', 'nw']
steps = []
word = ''
for line in Lines:
    step = []
    for char in line[:-1]:
        word += char
        if word in directions:
            step.append(word)
            word = ''
    steps.append(step)
    
print(steps)
              
```

```python
# calculate coordinates of each step
coordinations = []
for step in steps:
    coordination = [0,0,0]
    for direction in step:
        if direction == 'e':
            coordination[0] += 1
        elif direction == 'w':
            coordination[0] -= 1
        elif direction == 'sw':
            coordination[1] += 1
        elif direction == 'ne':
            coordination[1] -= 1
        elif direction == 'nw':
            coordination[2] += 1
        elif direction == 'se':
            coordination[2] -= 1
    coordinations.append(coordination)
print(len(coordinations))
```

```python
unique_coordinations = [list(c) for c in set(tuple(c) for c in coordinations)]
len(unique_coordinations)
```

```python
counts = dict((c, coordinations.count(list(c))) for c in set(tuple(c) for c in coordinations))
result = 0
for count in counts.values():
    if count%2 !=0:
        result +=1
result
```
