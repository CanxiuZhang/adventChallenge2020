```python
# read input file

input = open('./input.txt', 'r') 
Lines = input.readlines() 
  
count = 0
for line in Lines: 
    print("Line{}: {}".format(count, line.strip())) 
    count += 1
```

```python
# clean data
directions = ['e', 'w', 'se', 'sw', 'ne', 'nw']
steps = []
word = ''
for line in Lines:
    step = []
    for char in line:
#         print('char', char)
        word += char
#         print('word', word)
        if word in directions:
            step.append(word)
            word = ''
    word = ''
    steps.append(step)
    
print(steps)
              
```

```python
# calculate coordinates of each step
coordinations = []
for step in steps:
    coordination = [0,0]
#     print(step)
    for direction in step:
        if direction == 'e':
            coordination[0] += 1
        elif direction == 'w':
            coordination[0] -= 1
        elif direction == 'sw':
            coordination[0] -= 0.5
            coordination[1] -= 0.5
        elif direction == 'ne':
            coordination[0] += 0.5
            coordination[1] += 0.5
        elif direction == 'nw':
            coordination[0] -= 0.5
            coordination[1] += 0.5
        elif direction == 'se':
            coordination[0] += 0.5
            coordination[1] -= 0.5
        else:
            print('wrong char')
    coordinations.append(coordination)
print(coordinations)
```

```python
counts = dict((c, coordinations.count(list(c))) for c in set(tuple(c) for c in coordinations))
result = 0
for count in counts.values():
    if count%2 !=0:
        result +=1
result
```
