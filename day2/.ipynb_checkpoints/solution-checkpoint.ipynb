```python
# read input

input = open('./input.txt', 'r') 
Lines = input.readlines() 
  
count = 0
for line in Lines: 
    print("Line{}: {}".format(count, line.strip())) 
    count += 1
```

```python
# clean data
import re
regex = re.compile('\W+')
polices = []

for line in Lines:
    rule = regex.split(line);
    polices.append({'min': int(rule[0]), 'max': int(rule[1]), 'letter':rule[2], 'password': rule[3]})
print(len(polices))
print(len(Lines))
```

```python
# q1: count invalid password
result = 0
for police in polices:
    count = police['password'].count(police['letter'])
    if (count > police['max'] or count < police['min']):
        result += 1
        
print(result)
1000-458
```
