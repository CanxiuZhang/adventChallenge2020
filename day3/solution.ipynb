```python
# read input file

input = open('./input.txt', 'r') 
Lines = input.readlines() 
  
count = 0
for line in Lines: 
    print("Line{}: {}".format(count, line.strip())) 
    count += 1
```

```python
# clean data
def split(string):
    return [char for char in string]

patterns = []
for line in Lines:
    patterns.append(split(line)[:-1])
    
print(len(patterns[0]))
```

```python
# Q1 count tree
count = 0
row = 0
pattern_length = len(patterns[0])
for pattern in patterns:
    index = (row*3)%pattern_length
#     print('index', index)
    if pattern[index] == '#':
        count += 1
    row += 1
print('count', count)
```

```python
# Q2 different combination
def countTreesByPattern(right, down, patterns):
    count = 0
    row = 0
    pattern_length = len(patterns[0])
    time = 0
    
    for pattern in patterns:
        if row%down == 0:
            index = (time*right)%pattern_length
#             print("row: {:d} index: {:d}".format(row, index))
            if pattern[index] == '#':
                count += 1
            time += 1
        row += 1
    return count
        
print(countTreesByPattern(1,2,patterns))    
```

```python
countTreesByPattern(3,1,patterns) * countTreesByPattern(1,1,patterns) * countTreesByPattern(5,1,patterns) * countTreesByPattern(7,1,patterns) * countTreesByPattern(1,2,patterns)
```
